```python
def calculate_sub_index(value, pollutant):
    for bp in breakpoints[pollutant]:
        if bp['bp_low'] <= value <= bp['bp_high']:
            return ((bp['aqi_high']-bp['aqi_low'])/(bp['bp_high']-bp['bp_low']))*(value-bp['bp_low']) + bp['aqi_low']
    return None
        
def calculate_aqi(row):
    sub_indices = []
    for pollutant in breakpoints.keys():
        sub_index = calculate_sub_index(row[pollutant], pollutant)
        if sub_index is not None:
            sub_indices.append(sub_index)
    return max(sub_indices) if sub_indices else None
```

```python
df['AQI'] = df.apply(calculate_aqi, axis=1)
```
